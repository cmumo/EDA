```python
print(df['Salary'].quantile(0.25))
print(df['Salary'].quantile(0.75))
```

```python
q1 = df['Salary'].quantile(0.25)
q3 = df['Salary'].quantile(0.75)
iqr = q3 - q1

lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr
outliers = df[(df['Salary'] < lower_bound) | (df['Salary'] > upper_bound)]

print('Number of outliers:', len(outliers))
print(outliers['Salary'])
```

```python
# remove outliers
df = df[(df['Salary'] >= lower_bound) & (df['Salary'] <= upper_bound)]
print('Number of outliers removed:', len(outliers))
print('New shape of DataFrame:', df.shape)
```

```python
df['Salary'].plot(kind='box')
```

```python
average_salary_by_collegetier = df.groupby('CollegeTier')['Salary'].mean()
average_salary_by_collegetier.plot(kind='bar', figsize=(10,6))

plt.title('Average Salary by College Tier')
plt.xlabel('College Tier')
plt.ylabel('Average Salary')

plt.show()
```

Students from Tier1 colleges have a higher salary compared to students from tier 2 colleges

```python
average_salary_by_graduationYear = df.groupby('GraduationYear')['Salary'].mean()
average_salary_by_graduationYear.plot(kind='bar', figsize=(10,6))

plt.title('Average Salary by Graduation Year')
plt.xlabel('Year')
plt.ylabel('Average Salary')

plt.show()
```

There seems to be no relation between the salary offered with the year of graduation
